```python
# ordenação de valores
usuarios = ['Pablo', 'Anna', 'Raissa']
print(usuarios)
idades = [15, 87, 32, 65, 56, 32, 49, 37]
print(idades)
```

```python
print(f"Ordem crescente: {sorted(idades)}")
print(f"Ordem decrescente: {reversed(idades)}")
print(f"Ordem decrescente: {list(reversed(idades))}")
print(f"Ordem decrescente: {sorted(idades, reverse=True)}")
print(f"Ordem decrescente: {list(reversed(sorted(idades)))}")
```
