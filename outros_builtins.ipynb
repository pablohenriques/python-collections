```python
# percorrendo sequências e verificando seus índices

idades = [15, 87, 32, 65, 56, 32, 49, 37]

for i in range(len(idades)):
    print(i, idades[i])
```

```python
# lazy, gera uma sequência de forma gradual e não de forma complata
print(range(len(idades)))
print(enumerate(idades))

# mas é possível forçar uma sequência de forma completa, com list
print(list(range(len(idades))))
print(list(enumerate(idades)))

# lazy
for valor in enumerate(idades):
    print(valor)

# desempacotamento
for indice, valor in enumerate(idades):
    print(f"Índice: {indice} - Valor: {valor}")
```
