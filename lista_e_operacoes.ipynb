Compreendo coleções em Python

As lista em Python representam uma sequência de valores. Especificamente no Python, o tipo de sequência lista pode receber valores de diversos tipos. O tipo lista, oferece por padrão algumas funcionalidades que permitem realizar diversas operações. Abaixo segue lista de operações. 

```python
idades = [10, 20, 30, 40, 50] # inicializando uma lista de idades
print(f"Valores de idades: {idades}")
```

```python
print(f"Tipo de idades: {type(idades)}")
```

```python
idades.append(15) # adicionando uma idades ao final da lista de idades
print(f"Nova lista de idades")
print(f"Verificando a quantidade de valores na lista idades: {len(idades)}")
idades.remove(15) # removendo valor 15 da lista de idades
print(f"Valores em lista idades: {idades}")
idades.clear() # limpando lista
print(f"Valores em lista de idades: {idades}")
```

```python
idades = [10, 20, 30, 40, 50] # inicialização de lista
for idade in idades:
    print(f"Idade: {idade}")
print(f"\nVerificando valores na lista")
print(f"28 está na lista: {28 in idades}")
print(f"40 está na lista: {40 in idades}")
idades.insert(0, 20) # inserindo um novo valor de forma indexada
print(f"Valores em lista: {idades}")
```

Valores mutáveis, em Python, possuem uma consequência de serem alterado em qualquer momento na execução de um código, depedendo do contexto é ideal utilizar lista imutáveis sempre que possível parar evitar surpresas. Por exemplo, ao passar uma lista para uma função, perdemos o controle sobre a lista e nestes casos, devem ser avaliadas soluções, como a imutabilidade de sequências.

```python
def listar_valores(lista):
    print(f"Lista de valores: {lista}")
    lista.append(13)

idades = [10, 20, 30]
print(f"Idades: {idades}")
listar_valores(idades)
print(f"Idades (sequência alterada): {idades}")
```

O "erro" citado como um consequência de lista está na referência, tanto o parâmetro da função listar_valores (lista) quanto idades apontam para o mesmo local na memória e por conta da natureza da lista (mutável) os valores são alterados, o que pode causar problemas.

Tratando-se de referência é necessáio atenção, pois mesmo um parâmtro em uma função, pode referenciar uma espaço na memória causando "anomalias" no desenvolvimento.

```python
# ERRADO, este parâmetro não será considerado opicional, pois 'lista' estará apontando para um espaço na memória
def minha_lista(lista = []):
    print(lista)
    lista.append(1)

minha_lista()
minha_lista()
minha_lista()
```

```python
# CERTO, agora, referênciando None, o parâmetro 'lista' será um parâmetro opcional
def minha_lista(lista = None):
    if lista is None:
        lista = list()
    print(lista)
    lista.append(1)

minha_lista()
minha_lista()
minha_lista()
```

```python
# list comprehension
lista = [numero for numero in range(10)]
print(f"Lista: {lista}")
lista2 = [numero for numero in range(10) if numero % 2 == 0]
print(f"Lista2: {lista2}")
```
