```python
# Conjuntos em Python são mutáveis, é possível alterar e remover valores de um conjunto

usuarios = {1, 5, 76, 34, 63, 52, 19, 13, 17}
len(usuarios)
```

```python
usuarios.add(13)
print(len(usuarios))
usuarios.add(139)
print(len(usuarios))
```

```python
# Contudo, é possível deixar um conjunto imutável, congelando ele. Não sendo possível adicionar valores em um conjunto congelado.
usuarios = frozenset(usuarios)
print(f"Conjunto (congelado) de usuários: {usuarios} - Tipo: {type(usuarios)}")
```
