```python
# acessando valores em um diconário
meu_dicionario = {"Nome": 3, "Cachorro": 2, "Gosto": 5}
print(f"{meu_dicionario['Nome']}")
# print(f"{meu_dicionario['Voz']}") aqui ocorre um erro

print(f"{meu_dicionario.get('voz', 0)}") # retorna 0 caso o valor não exista
print(f"{meu_dicionario.get('Nome', 0)}")
```

```python
# adicionando um valor no dicionário
meu_dicionario["Maria"] = 2
print(meu_dicionario)

# alterando valor em uma chave no meu dicionário
meu_dicionario["Maria"] = 3
print(meu_dicionario)

# removendo um elemento do dicionário
del meu_dicionario["Maria"]
print(meu_dicionario)

# verificar a existência de um elemento em um dicionário (a verificação é feita nas chaves do dict)
print(f'Nome existe em dicionários: {"Nome" in meu_dicionario}')

# iteração entre os elementos do dicionário (chaves)
for elemento in meu_dicionario:
    print(f"Elemento: {elemento}")

# iteração entre os elementos do dicionário (chaves)
for elemento in meu_dicionario.keys():
    print(f"Chave: {elemento}")

# iteração entre os elementos do dicionário (valores)
for elemento in meu_dicionario.values():
    print(f"Valor: {elemento}")

# iteração entre os elementos do dicionário (linhas)
for elemento in meu_dicionario.items():
    print(f"Elemento: {elemento}")

# iteração entre os elementos do dicionário (linhas - desempacotando)
for chave, valor in meu_dicionario.items():
    print(f"{chave}: {valor}")
```
