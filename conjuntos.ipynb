```python
usuarios_data_science = [15, 23, 43, 56]
usuarios_machine_learning = [13, 23, 56, 42]
```

```python
# cópia rasa das listas
assistiram = usuarios_data_science.copy()
assistiram.extend(usuarios_machine_learning)
print(f"Usuários de assitiram: {assistiram}")
print(f"Total de usuários: {len(assistiram)}")
```

```python
# conjunts em Python
print(f"Conjunto de usuários: {set(assistiram)}")

# criando um conjunto
print(set([1, 2, 3, 1]))
print(f"Tipo: {type(set())}")
```

```python
for u in set(assistiram):
    print(u)
```

```python
# uniao de conjuntos
set(usuarios_data_science) | set(usuarios_machine_learning)
```

```python
# intersecção de conjuntos
set(usuarios_data_science) & set(usuarios_machine_learning)
```

```python
# subtração de conjuntos
set(usuarios_data_science) - set(usuarios_machine_learning)
```

```python
# verificando se valor existe em um conjunto
ds_not_ml = set(usuarios_data_science) - set(usuarios_machine_learning)
15 in ds_not_ml
```

```python
# 'ou' exclusivo
set(usuarios_data_science) ^ set(usuarios_machine_learning)
```
